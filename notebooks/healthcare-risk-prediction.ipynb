# Healthcare Risk Prediction System using Machine Learning

## 1. Introduction

Chronic diseases such as stroke, diabetes, and hypertension are among the leading contributors to global mortality and long-term disability. Early identification of individuals at risk plays a crucial role in preventive healthcare and clinical decision support systems.

In this project, we develop a machine learning-based multi-disease risk prediction framework using structured health survey data (BRFSS-based features). The system is designed to:

- Build and evaluate predictive models for three major chronic conditions
- Compare multiple supervised learning algorithms under a unified evaluation framework
- Analyze model performance under real-world conditions such as class imbalance and heterogeneous feature distributions
- Compare model performance using multiple evaluation metrics to understand disease-specific predictive behavior

Each disease is modeled as an independent binary classification task within a standardized machine learning pipeline. This **comparative modeling framework** enables systematic evaluation of how different supervised learning algorithms perform across multiple clinical prediction tasks.

The final objective is to derive a robust, interpretable, and reproducible ML pipeline suitable for real-world healthcare risk stratification.

# 4. Disease-specific Prediction Pipelines

Each disease is treated as an independent binary classification task following a standardized machine learning workflow. For every disease prediction pipeline, the following steps are performed:

- Data preparation and feature scaling
- Model training using multiple machine learning algorithms
- Model evaluation using performance metrics
- Comparative analysis of model performance using multiple evaluation metric

## 4.1 Stroke Prediction

### Data Preparation

```python
# Split Features (X) and Target Variable (y)

X = df.drop("Stroke", axis=1)
y = df["Stroke"]

print("Feature Matrix Shape:", X.shape)
print("Target Vector Shape:", y.shape, "\n")


# Split the Dataset into Training and Testing Sets

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=105,
    stratify=y
)


# Display the Dimensions of the Training and Testing Sets

print("Training Features :", X_train.shape)
print("Testing Features  :", X_test.shape)
print("Training Labels   :", y_train.shape)
print("Testing Labels    :", y_test.shape)
```

```python
# Feature Scaling

scaler = StandardScaler()

X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

### Shared Model Utilities

```python
# Initialize Model Accuracy Storage
model_accuracies = {
    "Stroke": {},
    "Diabetes": {},
    "Hypertension": {}
}


# Initialize Model Performance Storage
model_metrics = {
    "Stroke": [],
    "Diabetes": [],
    "Hypertension": []
}
```

```python
# Train, Evaluate, and Compare Machine Learning Models

def model_train_test(disease, model_name, model):

    # Train the model
    model.fit(X_train_scaled, y_train)

    # Make predictions
    y_pred = model.predict(X_test_scaled)

    # Accuracy
    accuracy = accuracy_score(y_test, y_pred) * 100    
    model_accuracies[disease][model_name] = accuracy
    print(f"Accuracy: {accuracy:.2f}%\n")



    # Classification report
    report = classification_report(
        y_test,
        y_pred,
        output_dict=True,
        zero_division=0
    )
    #print(report.keys())
    print(classification_report(y_test, y_pred, zero_division=0))

    model_metrics[disease].append({
        "Model": model_name,
        "Accuracy": accuracy,
        "Precision": report["1.0"]["precision"] * 100,
        "Recall": report["1.0"]["recall"] * 100,
        "F1-score": report["1.0"]["f1-score"] * 100
        })



    # Confusion matrix
    cm = metrics.confusion_matrix(y_test, y_pred)
    fig, ax = plot_confusion_matrix(
        conf_mat=cm,
        figsize=(6, 6),
        cmap="GnBu"
    )

    plt.title(f"{model_name} - Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()



    # ROC curve
    RocCurveDisplay.from_estimator(
        model,
        X_test_scaled,
        y_test
    )

    plt.title(f"{model_name} - ROC Curve")
    plt.show()
```

```python
# Plot Model Accuracy Comparison

def plot_bar(disease):

    models = [ "Logistic Regression", "Gaussian NB", "Bernoulli NB", "KNN", "Decision Tree" ]
    accuracies = list(model_accuracies[disease].values())

    plt.figure(figsize=(9, 5))
    plt.bar(models, accuracies, color="#4C78A8")

    plt.title(f"{disease} - Model Accuracy Comparison", fontsize=15)
    plt.xlabel("Machine Learning Models", fontsize=10)
    plt.ylabel("Accuracy (%)", fontsize=10)
    plt.ylim(0, 100)
    plt.show()
```

```python
# Display Model Performance Summary
 
def performance_summary(disease):

    summary = pd.DataFrame(model_metrics[disease])
    summary = summary.sort_values(by="Accuracy", ascending=False)

    display(
        summary.style.format({
            "Accuracy": "{:.2f}%",
            "Precision": "{:.2f}%",
            "Recall": "{:.2f}%",
            "F1-score": "{:.2f}%"
        })
    )
```

### Model Training

**LOGISTIC REGRESSION**

```python
# Logistic Regression with Hyperparameter Tuning

param_grid = {
    "C": np.logspace(-3, 3, 7),
    "penalty": ["l1", "l2"]
}

logistic_model = GridSearchCV(
    estimator=LogisticRegression(
        solver="liblinear",
        random_state=42,
        class_weight="balanced"
    ),
    param_grid=param_grid,
    cv=10,
    scoring="accuracy",
    n_jobs=-1
)

model_train_test(
    disease="Stroke",
    model_name="Logistic Regression",
    model=logistic_model
)
```

**GAUSSIAN NAIVE BAYES**

```python
# Gaussian Naive Bayes

gaussian_nb = GaussianNB()

model_train_test(
    disease="Stroke",
    model_name="Gaussian Naive Bayes",
    model=gaussian_nb
)
```

**BERNOULLI NAIVE BAYES**

```python
# Bernoulli Naive Bayes

bernoulli_nb = BernoulliNB()

model_train_test(
    disease="Stroke",
    model_name="Bernoulli Naive Bayes",
    model=bernoulli_nb
)
```

**K-Nearest Neighbors (KNN)**

```python
# K-Nearest Neighbors with Hyperparameter Tuning

param_grid = {
    "n_neighbors": range(1, 20)
}

knn_search = GridSearchCV(
    estimator=KNeighborsClassifier(),
    param_grid=param_grid,
    cv=10,
    scoring="accuracy",
    n_jobs=-1
)

knn_search.fit(X_train_scaled, y_train)
```

```python
# Best Hyperparameter

print("Best Parameters:", knn_search.best_params_)

best_cv_accuracy = knn_search.best_score_ * 100
print(f"Best Cross-Validation Accuracy: {best_cv_accuracy:.2f}%")
```

```python
# K-Nearest Neighbors

best_knn = knn_search.best_estimator_

model_train_test(
    disease="Stroke",
    model_name="KNN",
    model=best_knn
)
```

**DECISION TREE**

```python
# Decision Tree with Hyperparameter Tuning

param_grid = {
    "criterion": ["gini", "entropy"],
    "max_depth": [5, 6, 7, 8, 9],
    "max_features": ["sqrt", "log2"],
    "ccp_alpha": [0.1, 0.01, 0.001]
}

decision_tree_search = GridSearchCV(
    estimator=DecisionTreeClassifier(random_state=42),
    param_grid=param_grid,
    cv=5,
    scoring="accuracy",
    n_jobs=-1
)

decision_tree_search.fit(X_train_scaled, y_train)
```

```python
# Best Hyperparameters

print("Best Parameters:", decision_tree_search.best_params_)

best_cv_accuracy = decision_tree_search.best_score_ * 100

print(f"Best Cross-Validation Accuracy: {best_cv_accuracy:.2f}%")
```

```python
# Decision Tree

best_decision_tree = decision_tree_search.best_estimator_

model_train_test(
    disease="Stroke",
    model_name="Decision Tree",
    model=best_decision_tree
)
```

### Model Evaluation

```python
plot_bar('Stroke')
```

```python
performance_summary("Stroke")
```

**MODEL INTERPRETATION**

Although **Decision Tree** achieved the highest accuracy (**93.78%**), it failed to identify positive stroke cases (Recall: **0.00%**), indicating poor performance on the minority class. 

In contrast, **Logistic Regression** achieved the highest recall (**73.09%**), making it more suitable for identifying individuals at risk despite its lower overall accuracy.
